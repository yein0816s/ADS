```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# Load the dataset
GSS = pd.read_csv('trends-gss.csv')

# Recodes
#Change the new variable name to appropriate one
GSS['losejob'] = np.where(GSS['joblose'] <= 2, 1, 0)
GSS['jobsat'] = np.where(GSS['satjob'] <= 2, 1, 0)
GSS['white'] = np.where(GSS['race'] == 1, 1, 0)
GSS['nyear'] = GSS['year'] - 1973
```

```python
losejob80 = smf.ols('losejob ~ jobsat + age + sex + educ + white + nyear', data=GSS[(GSS['year'] > 1980) & (GSS['year'] <= 1990)]).fit()
losejob90 = smf.ols('losejob ~ jobsat + age + sex + educ + white + nyear', data=GSS[(GSS['year'] > 1990) & (GSS['year'] <= 2000)]).fit()
losejob00 = smf.ols('losejob ~ jobsat + age + sex + educ + white + nyear', data=GSS[(GSS['year'] > 2000) & (GSS['year'] <= 2010)]).fit()

# Display regressions
print(losejob80.summary())
```

```python
# Visualize the trend
by_year = GSS.groupby('year')['losejob'].mean().reset_index()
plt.plot(by_year['year'], by_year['losejob'], color='blue', linewidth=2)
plt.xlabel('Year')
plt.ylabel('Mean')
plt.title('Trend of Mean "losejob" by Year')
plt.show()
```

```python
# Regressions
losejob80 = smf.ols(formula='losejob ~ jobsat + age + C(sex) + educ + white + nyear',
                data=GSS[(GSS['year'] > 1980) & (GSS['year'] <= 1990)])
results80 = losejob80.fit()

losejob90 = smf.ols(formula='losejob ~ jobsat + age + C(sex) + educ + white + nyear',
                data=GSS[(GSS['year'] > 1990) & (GSS['year'] <= 2000)])
results90 = losejob90.fit()

losejob00 = smf.ols(formula='losejob ~ jobsat + age + C(sex) + educ + white + nyear',
                data=GSS[(GSS['year'] > 2000) & (GSS['year'] <= 2010)])
results00 = losejob00.fit()

# Display regressions
print(results80.summary())
print(results90.summary())
print(results00.summary())
```

```python
# Visualize the trend
by_year = GSS.groupby('year')['losejob'].mean()

plt.plot(by_year.index, by_year.values, color='blue', linewidth=2)
plt.xlabel('Year')
plt.ylabel('Mean')
plt.show()
```

```python
from scipy.stats import norm
# Compare coefficients
coefs = pd.DataFrame({'e80': results80.params,
                      'se80': results80.bse,
                      'e90': results90.params,
                      'se90': results90.bse,
                      'e00': results00.params,
                      'se00': results00.bse})

#The difference in coefficients
coefs['b1minusb28090'] = coefs['e80'] - coefs['e90']

#standard error of the difference in coefficients 
coefs['denom8090'] = np.sqrt(coefs['se80'] ** 2 + coefs['se90'] ** 2)
#Calculate Z-scores
coefs['Z8090'] = coefs['b1minusb28090'] / coefs['denom8090']
#Calculate p-values
coefs['pvalue8090'] = 2 * (1 - norm.cdf(np.abs(coefs['Z8090'])))

coefs['b1minusb29000'] = coefs['e90'] - coefs['e00']
coefs['denom9000'] = np.sqrt(coefs['se90'] ** 2 + coefs['se00'] ** 2)
coefs['Z9000'] = coefs['b1minusb29000'] / coefs['denom9000']
coefs['pvalue9000'] = 2 * (1 - norm.cdf(np.abs(coefs['Z9000'])))

coefs['b1minusb28000'] = coefs['e80'] - coefs['e00']
coefs['denom8000'] = np.sqrt(coefs['se80'] ** 2 + coefs['se00'] ** 2)
coefs['Z8000'] = coefs['b1minusb28000'] / coefs['denom8000']
coefs['pvalue8000'] = 2 * (1 - norm.cdf(np.abs(coefs['Z8000'])))

# Display the final dataset
print(coefs)
```
